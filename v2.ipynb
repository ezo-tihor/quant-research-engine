```python
import pandas as pd
import numpy as np
import data.data_loader as data_loader
import backtester.backtester as backtester
import performance.analyzer as perf_analyzer
import strategy.equal_weight as equal_weight
```

```python
tickers = ['AAPL', 'MSFT', 'GOOG']
# tickers = ['AAPL']
prices = data_loader.load_price_data(tickers)
```

```python

print(prices.head())
print(prices.shape)
```

```python
strategy = equal_weight.EqualWeightStrategy()
```

```python
bt = backtester.Backtester(data = prices, strategy = strategy)
results = bt.run()
```

```python
for key, value in results.items():
    print(key, results[key].head(3), sep='\n')
```

```python
pa = perf_analyzer.PerformanceAnalyzer(results)
```

```python
summary = pa.summary()
```

```python
for key, value in summary.items():
    print(key, value, sep=': ')
```

```python
def log_results(name, summary):
    print(f"\n=== {name} ===")
    for k,v in summary.items():
        print(k,v, sep=": ")
```

```python
log_results('first log', summary)
```

```python
def save_results(name, summary, filename='results.csv'):
    df = pd.DataFrame([summary])
    df['strategy'] = name
    
    df.to_csv(filename, index=False)
```

```python
save_results('first logging', summary)
```
